# Fig 8: Plot Bacon-Shor levels

The notebook generates Fig.8 in the paper

```python
import numpy as np
import matplotlib.pyplot as plt
from typing import List, Dict
import matplotlib
matplotlib.rcParams['mathtext.fontset'] = 'cm'
import json
```

```python
tex_rc_params = {
    'backend': 'ps',
    'axes.labelsize': 10,
    'axes.titlesize': 10,
    'legend.fontsize': 10,
    'xtick.labelsize': 8,
    'ytick.labelsize': 8,
    'font.family': 'serif',
}

%matplotlib inline
%config InlineBackend.figure_format='retina'  # for hi-dpi displays
```

```python
with open('bacon_shor_levels.txt') as file_data:
    result = json.load(file_data)
```

```python
def plot_bs_levels(
    result: Dict,
    tex_rc_params: dict,
    save=False
):
    with plt.rc_context(tex_rc_params):
        fig, ax = plt.subplots( 
            figsize=(3.4, 2.6), 
            constrained_layout=True
        )
        matplotlib.rcParams['mathtext.fontset'] = 'cm'
        phi_x = result["phi_x"]
        
        cmap_blue = matplotlib.cm.get_cmap('Blues')
        cmap_red = matplotlib.cm.get_cmap('Reds')
        
        #colors={"sz0": cmap_blue(1.0), "sz1": cmap_red(0.8)}
        
        colors={"sz0": 'black', "sz1": 'black'}
        
        linestyles = {"sz0": 'solid', 'sz1': 'dashed'}
        
        #ax.plot(phi_x, np.zeros(len(phi_x)), color='gray', linestyle=':')
        
        for lev in range(0, 3):
            energies_sz0 = [result["sz0"][x][lev] for x in range(len(result["sz0"]))]
            energies_sz1 = [result["sz1"][x][lev] for x in range(len(result["sz1"]))]
            ax.plot(phi_x, energies_sz0, linestyle=linestyles["sz0"], color=colors["sz0"])
            ax.plot(phi_x, energies_sz1, linestyle=linestyles["sz1"], color=colors["sz1"])
        ax.set_xlabel("$\phi_x$")
        ax.set_ylabel("$\mathrm{Energy}$")
        x_ticks = [0, np.pi/2, np.pi, 3/2*np.pi, 2*np.pi]
        x_ticks_latex = ['$0$', '$\pi/2$', '$\pi$', '$3 \pi/2$', '$2 \pi$']
        ax.set_xticks(x_ticks)
        ax.set_xticklabels(x_ticks_latex)
        ax.set_ylim(-1.0, 6.0)
        y_ticks = [-1, 0, 1, 2, 3, 4, 5, 6]
        y_ticks_latex = ['$-1$', '$0$', '$1$', '$2$', '$3$', '$4$', '$5$', '$6$']
        ax.set_yticks(y_ticks)
        ax.set_yticklabels(y_ticks_latex)
        if save:
            plt.savefig("bacon_shor_levels.svg", bbox_inches='tight', 
                        transparent=True,
                        pad_inches=0)
            plt.savefig("bacon_shor_levels.pdf", bbox_inches='tight', 
                        transparent=True,
                        pad_inches=0)
        plt.show()
        
```

```python
plot_bs_levels(result, tex_rc_params, save=True)
```
